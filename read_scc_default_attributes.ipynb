# Helper code to read the defaults of NREL's SCC defaults
They're defined "in name" here, but you can't actually find the values those defaults set anywhere.
https://nrel-pysam.readthedocs.io/en/main/modules/Battery.html#


```python
import marshal
import pprint
import os

directory = "/opt/homebrew/lib/python3.11/site-packages/PySAM/defaults/"

files = [filename for filename in os.listdir(directory)
         if os.path.isfile(os.path.join(directory, filename))]
for filename in sorted(files):
    print(filename)

selected_file = "battery_genericbatteryresidential"
```

```python

df_file = f"/opt/homebrew/lib/python3.11/site-packages/PySAM/defaults/{selected_file}.df"

def print_keys(d, indent=0):
    if isinstance(d, dict):
        for key, value in d.items():
            print("  " * indent + str(key))
            print_keys(value, indent + 1)

with open(df_file, "rb") as f:
    data = marshal.load(f)

print_keys(data)
```

```python
group_name = "Load"
property = "grid_outage"

value = data[group_name][property]
print(f"{group_name}.{property}:", value)
```
